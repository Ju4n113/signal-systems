## Sampling

Muestrear (o sampling) una señal electrica desde el dominio continuo es adquirir valores repetidamente a una tasa constante y rapida con la intension de no perder valores.

Definimos frecuencias de sampling ($f_s$) al número de muestras capturadas por segundo, la inversa se define como periodo de muestreo ($t_s = 1/f_s$). 

El valor de cada muestra posee resolucion finita y esta dada por el "numero de bits" de la conversion analogica a digital (tipicamente entre 16 y 24 bits, para mantener una relacion señal ruido > 96 dB).

Por consiguiente, en un gráfico de amplitud y tiempo, el waveform de una señal analogica (continua) se aproxima mediante una secuencia de puntos (discretos), ya que tanto el tiempo como la amplitud son múltiplos enteros de pequeñas "unidades de muestreo" de tiempo y amplitud.

Pensemos en una secuencia de puntos $[x_1,x_2,...,x_n]$ y luego se nos dice que reconstruyamos la función. ¿Qué haríamos? Este es el tipo de pregunta que rara vez se hace porque típicamente tomamos muestras, filtramos y luego hacemos otra cosa sin intentar reconstruir la función directamente desde las muestras. Quizás lo más natural que se puede hacer es trazar una línea recta entre cada punto como en la interpolación lineal. Como se oberva en el siguiente script

```python
import pylab as# libreria de comandos "muy parecidos" a MATLAB PLOT.
import numpy as np

# Hz, signal frequency
%pylab inline 
```

```python
fig,ax = subplots()

f = 1.0  # Hz, frecuencia de la señal
fs = 20.0 # Hz, frecuencia de muestreo (min > 2*f) 
t = np.arange(-1,1+1/fs,1/fs) # array de intervalosample interval, symmetric for convenience later
x = np.sin(2*np.pi*f*t)
ax.plot(t,x,'o-')
ax.set_xlabel('Tiempo',fontsize=10)
ax.set_ylabel('Amplitud',fontsize=10)
ax.axis( xmin = -1, xmax = 1, ymin = -1, ymax = 1)
ax.set_title('Muestreo de señal continua',fontsize=10)
# fig.savefig('figure_00@.png', bbox_inches='tight', dpi=300)
```

De la anterior representación surge la siguiente pregunta ¿Puede una muestra digital representar fielmente una señal analogica?

Si, pero solo siguiendo el teorema "Nyquist-Shannon", una definición valida es:

*"Si una función $x(t)$ no contiene frecuencias mayores a B Hz, ésta es completamente determinada por la obtención ordenada de una serie de puntos 1/2B segundos espaciados."*

La función $x(t)$ es una funcion de $\mathbb{R}\to\mathbb{R}$ con infinitos puntos entre dos ordenadas, por lo cual al muestreo se lo puede ver como una reduccion masiva de datos, ya que solo se necesita una pequeña cantidad de puntos para caracterizar por completo la función. Esta es una idea poderosa que vale la pena explorar. De hecho, hemos visto esta idea de reducir una función a un conjunto discreto de números antes en expansiones de series de Fourier para $x(t)$  periodica. Recordemos:

Análisis: $\frac{1}{T}\int_{0}^{T} x(t)exp(-jw_nt)dt$

Sintesis: $\sum_{k}^{} a_n exp(jwt)$

Pero aca estamos generando puntos discretos $a_n$ e integrando toda la funcion $x(t)$,no evaluando en un solo punto. Esto significa que estamos recopilando información sobre la función completa para calcular un único punto discreto y $a_n$, mientras que con el muestreo solo estamos tomando puntos individuales de forma aislada.

En esta trama, observe cuán cerca de los extremos del $sin$ en $t=1/(4f)$ y $t=3/(4f)$, estamos tomando la misma densidad de puntos ya que el teorema de muestreo no exige que muestremos a intervalos regulares. Esto significa que en las pendientes ascendentes y descendentes de la $sin$, que son evidentemente lineales y donde una aproximación lineal es buena, estamos tomando la misma densidad de muestras que cerca de las concavidades. el siguiente lo muestra:

```python
fig,ax = subplots()

ax.plot(t,x,'o-')
ax.axis( xmin = 1/(4*f)-1/fs*3, xmax = 1/(4*f)+1/fs*3, ymin = 0, ymax = 1.1 )
ax.set_xlabel('Tiempo',fontsize=10)
ax.set_ylabel('Amplitud',fontsize=10)
ax.set_title('Zoom en una zona de la señal',fontsize=10)
# fig.savefig('figure_00@.png', bbox_inches='tight', dpi=300)
```

Vamos a realizar la reconstruccion de la interpolacion lineal por partes y comparar la calidad de la aproximacion usando numpy.piecewise:

```python
interval=[] # dominio de piecewise
apprx = []  # dominio line
# acumule puntos uniformemente dentro de intervalos
tp = hstack([ linspace(t[i],t[i+1],20,False) for i in range(len(t)-1) ])
# construccion de argumentos para piecewise2
for i in range(len(t)-1):
   interval.append( np.logical_and(t[i] <= tp,tp < t[i+1]))
   apprx.append( (x[i+1]-x[i])/(t[i+1]-t[i])*(tp[interval[-1]]-t[i]) + x[i])
x_hat = np.piecewise(tp,interval,apprx) # piecewise linear approximation
```

Ahora, podemos examinar el error cuadratico en el interpolador. El siguiente fragmento muestra los errores y con el error relleno del interpolador lineal.

```python
ax1 = figure().add_subplot(111)
ax1.fill_between(tp,x_hat,sin(2*pi*f*tp),facecolor='red')
ax1.set_xlabel('time',fontsize=10)
ax1.set_ylabel('Amplitude',fontsize=10)
ax2 = ax1.twinx()
sqe = ( x_hat - sin(2*pi*f*tp))**2
ax2.plot(tp, sqe,'r')
ax2.axis(xmin=-1,ymax= sqe.max() )
ax2.set_ylabel('error cuadratico', color='r',fontsize=18)
ax1.set_title('Errors with Piecewise Linear Interpolant')

# ax1.figure.savefig('figure_00@.png', bbox_inches='tight', dpi=300)
```

Nota: Le insto a cambiar la tasa de muestreo fs en el código anterior y luego vuelva a ejecutar este cuaderno para ver cómo estos errores cambian con más / menos puntos de muestreo.

Ahora bien, podríamos seguir esta línea de razonamiento con polinomios de orden superior en lugar de líneas rectas, pero todo esto finalmente nos llevaría a la misma conclusión; es decir, que todas estas aproximaciones mejoran a medida que aumenta la densidad de los puntos de muestra, que es exactamente lo opuesto a lo que dice el teorema de muestreo: existe un conjunto escaso de puntos de muestras que recuperarán la función original. Además, observamos que la calidad de la interpolación lineal por partes es sensible a donde se toman los puntos de muestra y el teorema de muestreo es tan poderoso que no tiene tal requerimiento.
